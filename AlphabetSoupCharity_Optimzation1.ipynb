```python
#Import dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import and read the charity_data.csv.
import pandas as pd 
app_df = pd.read_csv("../Resources/charity_data.csv")
app_df.head()
```

```python
#Dropping more or fewer columns.
app_df = app_df.drop(columns = ['EIN'])
```

I am not dropping anymore columns as previously as more data helps for a better accuarcy therefore I believe that keeping the name column might increase accuracy. 

```python
#Creating more bins for rare occurrences in columns.
#therfore will look at name
app_count = app_df['NAME'].value_counts()
app_count
```

```python
#Creating more bins for rare occurrences in columns.

# Choose a cutoff value and create a list of application types to be replaced
names_replace = list(app_count[app_count<10].index)
names_replace
```

```python
# Replace in dataframe
for name in names_replace:
    app_df['NAME'] = app_df['NAME'].replace(name,"Other")
# Check to make sure binning was successful
app_df['NAME'].value_counts()
```

```python
classification = app_df.CLASSIFICATION.value_counts()
classification
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(classification[classification < 500].index)

# Replace in dataframe
for cls in classifications_to_replace:
    app_df['CLASSIFICATION'] = app_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
app_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_dummies = pd.get_dummies(app_df)
```

```python
# Split our preprocessed data into our features and target arrays
X = app_dummies.drop(["IS_SUCCESSFUL"], axis='columns').values
y = app_dummies["IS_SUCCESSFUL"].values

```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

Test accuracy increases to a certain point with every addition of neurons ("wider" network) or layers ("deeper" network). At a point though it will start to overfit and accuracy will drop. 

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#Add more neurons to a hidden layer.
#Add more hidden layers.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  90
hidden_nodes_layer2 = 40
hidden_nodes_layer3= 20

dnn = tf.keras.models.Sequential()

# First hidden layer
dnn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
dnn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# third hidden layer
dnn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
dnn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
dnn.summary()
```

```python
# Compile the model
dnn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
dnn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = dnn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
